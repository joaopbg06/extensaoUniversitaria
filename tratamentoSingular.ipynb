```python
import pandas as pd
import numpy as np


ano = 20


df = pd.read_csv(f'./BasesLixoSP/coletaLixoSP_20{ano}.csv', encoding='ISO-8859-1', sep=';')


# sep=';' --> essencial para separar os valores, vulgo delimitador
# encoding='ISO-8859-1' --> precisa 'descriptografar' para ler corretamente os dados
```

```python
df.info()

colunas_converter = [f'{mes}/{ano}' for mes in ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']]

df.columns = df.columns.str.strip()



for mes_coluna in colunas_converter:
    df = df.fillna(0)
    df[mes_coluna] = df[mes_coluna].fillna('0').astype(str).str.strip().replace(r'^\s*$', '0', regex=True).str.replace(',', '.').str.replace('-', '0').astype(float)

df.rename(columns={"TIPO DE RESÖDUO": "Tipo_Residuo", 'ï»¿Tipo de resÃ­duo - Toneladas':'Tipo_Residuo',  'TIPO DE RESÍDUO': 'Tipo_Residuo', 'TIPO DE RESï¿½DUO': 'Tipo_Residuo', 'Tipo de Resíduo - Toneladas': 'Tipo_Residuo', 'Tipo de resï¿½duo - Toneladas': 'Tipo_Residuo', 'Tipo de resíduo - Toneladas': 'Tipo_Residuo', 'Tipo de residuo - Toneladas': 'Tipo_Residuo', 'TIPO DE RESDUO':'Tipo_Residuo', f"TOTAL 20{ano}": f"total_20{ano}", f'20{ano}': f'total_20{ano}'}, inplace=True)

df.columns = df.columns.str.lower().str.strip()
df['tipo_residuo'] = df['tipo_residuo'].str.replace('TOTAL', 'total geral')
df['tipo_residuo'] = df['tipo_residuo'].str.lower().str.title()

# tratamento de anos específicos
if ano >= 20 :
    df[f'total_20{ano}'] = df[colunas_converter].sum(axis=1)
else:
    df[f'total_20{ano}'] = df[f'total_20{ano}'].fillna('0').str.strip().replace(r'^\s*$', '0', regex=True).str.replace('.', '').str.replace(',', '.').str.replace('-', '0').astype(float)
    df[f'total_20{ano}'] = df[colunas_converter].sum(axis=1)


df['tipo_residuo'] = df['tipo_residuo'].str.strip()

if ano <= 18:
    df = df[df['tipo_residuo'] != 'Total Geral']
df = df[df['tipo_residuo'] != '(Vazio)']

if ano == 17:
    df = df.drop(columns='unnamed: 14')
if ano == 24:
    df = df.drop(columns='unnamed: 13')

    
display(df)
df.info()
```
