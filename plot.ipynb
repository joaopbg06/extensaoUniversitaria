```python
import pandas as pd
import matplotlib.pyplot as plt

df_soma = pd.read_excel('./DF/somas_total_2013_2024.xlsx')
df_tipos = pd.read_excel('./DF/tipo_residuos_total.xlsx')
```

```python
# Visualização
df_soma
```

```python
# Visualização
df_tipos
```

### Analises

```python
# 10 maiores tipos de residuo de todos os anos

df_filtro = df_tipos[df_tipos['tipo_residuo'] != 'total_geral']
total_anos_tipo = df_filtro.groupby('tipo_residuo')['total_anos'].sum().nlargest(10)

total_anos_tipo.plot(kind='barh')
```

```python
# Total de residuos por anos

total_por_ano = df_soma.groupby('ano')['soma_total'].sum()
total_por_ano.plot(kind='line')

```

```python
# Soma, media e mediana do total de residuos gerado

media = df_soma['soma_total'].mean().__round__(2)
soma_total = df_soma['soma_total'].sum().__round__(2)
mediana = df_soma['soma_total'].median().__round__(2)

print(f"A soma total é: {soma_total}")
print(f"A media é: {media}")
print(f"A mediana é: {mediana}")
```

```python
# Top 1 ano do total de residuo

maior_residuo = df_soma['soma_total'].max()
ano = df_soma['ano'][df_soma['soma_total'] == maior_residuo].iloc[0]

print(f'A maior ano registrado é {ano} com {maior_residuo}')

```

```python
# Grafico setor de 2013 e 2018 da distribuição de residuo

graf_setor1 = df_filtro.groupby('tipo_residuo')['total_2013'].sum().nlargest(5)
graf_setor2 = df_filtro.groupby('tipo_residuo')['total_2018'].sum().nlargest(5)


graf_setor1.plot(kind='pie')
plt.show()
graf_setor2.plot(kind='pie')
plt.show()

```
